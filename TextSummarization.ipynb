AI and ML Techniques for Cyber Security - Text Summerization

Imports

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string
import math
```

Tokenize Sentences

```python
text = """Artificial Intelligence is transforming many industries.
Machine learning helps computers learn from data.
AI is used in healthcare, finance, and automation."""

sentences = sent_tokenize(text)
```

Create Frequency Matrix

```python
print("Create Frequency Matrix")
stopWords = set(stopwords.words("english"))
freq_matrix = {}
for sent in sentences:
    freq_table = {}
    words = word_tokenize(sent.lower())
    #print("In Sentences loop")
    for word in words:
        #print("In Word loop:", word )
        if word in stopWords or word in string.punctuation:
            continue
        if word in freq_table:
            freq_table[word] += 1
        else:
            freq_table[word] = 1

    freq_matrix[sent[:15]] = freq_table
    #print(freq_matrix)
```

Calculate Term Frequency (TF)

```python
print("Calculate Term Frequency (TF)")
tf_matrix = {}
for sent, freq_table in freq_matrix.items():
    tf_table = {}
    total_words = sum(freq_table.values())
    #print("Total Words:", total_words)
    for word, count in freq_table.items():
        tf_table[word] = count / total_words

    tf_matrix[sent] = tf_table
    #print("TF Matrix:", tf_matrix)
```

Create Document Per Word Table

```python
word_per_doc = {}
print("Create Document Per Word Table")
for sent, freq_table in freq_matrix.items():
    for word in freq_table:
        if word in word_per_doc:
            word_per_doc[word] += 1
        else:
            word_per_doc[word] = 1
```

Calculate IDF Matrix

```python
print("Calculate IDF Matrix")
idf_matrix = {}
total_sentences = len(freq_matrix)

for sent, freq_table in freq_matrix.items():
    idf_table = {}
    
    for word in freq_table:
        idf_table[word] = math.log10(total_sentences / float(word_per_doc[word]))
        
    idf_matrix[sent] = idf_table
#print("IDF Matrix:", idf_matrix)
```

Calculate TF-IDF Matrix

```python
print("Calculate TF-IDF Matrix")
tf_idf_matrix = {}

for sent in tf_matrix:
    tf_idf_table = {}
    
    for word in tf_matrix[sent]:
        tf_idf_table[word] = tf_matrix[sent][word] * idf_matrix[sent][word]
        
    tf_idf_matrix[sent] = tf_idf_table
#print("TF-IDF Matrix:", tf_idf_matrix)
```

Score the Sentences

```python
print("Score the Sentences")
sentence_scores = {}
for sent, tfidf_table in tf_idf_matrix.items():
    total_score = sum(tfidf_table.values())
    sentence_scores[sent] = total_score / len(tfidf_table)
#print("Sentence Scores:", sentence_scores)
```

Find the Threshold

```python
print("Find the Threshold")
sum_scores = sum(sentence_scores.values())
threshold = sum_scores / len(sentence_scores)
#print("Threshold:", threshold)
```

Generate summary

```python
print("Generate summary")
summary = ""

for sent in sentences:
    if sentence_scores[sent[:15]] >= threshold:
        summary += " " + sent

print("Summary:-",summary)
```
